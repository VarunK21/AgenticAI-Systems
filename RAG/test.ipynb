```python
import pdfplumber

def extract_table_from_pdf(pdf_path):
    tables = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            table = page.extract_table()
            if table:
                tables.append(table)
    return tables

# Example Usage
pdf_path = "/workspaces/AgenticAI-Systems/RAG/Document/UNIT16_UNIT16.pdf"
tables = extract_table_from_pdf(pdf_path)

for i, table in enumerate(tables):
    print(f"Table {i+1}:")
    for row in table:
        print(row)
```
